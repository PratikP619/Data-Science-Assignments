**Q9-a  Calculate Skewness, Kurtosis & draw inferences on the following data.
      Cars speed and distance**

```python
# Loading the data set
df=pd.read_csv("/content/Q9_a.csv")
df
```

```python
df.head()
```

```python
# Calculate skewness and kurtosis for both columns

skewness=df[["speed","dist"]].skew()
kurtosis=df[["speed","dist"]].kurtosis()
```

```python
# Print results with  explanations
print("Skewness:\n", skewness)
print("kurtosis:\n", kurtosis)
```

```python
# Inferences
#A negative skewness value indicates that the distribution is left-skewed, meaning there are more outliers towards the lower values.
#A kurtosis value less than 3 indicates a light-tailed distribution, which means there are fewer outliers in both the lower and upper tails of the distribution.
```

**Q-11 Suppose we want to estimate the average weight of an adult male in    Mexico. We draw a random sample of 2,000 men from a population of 3,000,000 men and weigh them. We find that the average person in our sample weighs 200 pounds, and the standard deviation of the sample is 30 pounds. Calculate 94%,98%,96% confidence interval?**

```python
from scipy import stats
```

```python
# calculating
con_94=stats.t.interval(0.94,1999,loc=200,scale=30/np.sqrt(2000)) ## 0.94 is the confidence,n-1 which is DOF-1999
print(np.round(con_94,2))
```

```python
con_98=stats.t.interval(0.98,1999,loc=200,scale=30/np.sqrt(2000))
print(np.round(con_98,2))
```

```python
con_96=stats.t.interval(0.96,1999,loc=200,scale=30/np.sqrt(2000))
print(np.round(con_96,2))
```

**Q 20) Calculate probability from the given dataset for the below cases**


Calculate the probability of MPG  of Cars for the below cases.
       MPG <- Cars$MPG
a.	P(MPG>38)
b.	P(MPG<40)
c.    P (20<MPG<50)**

```python
q20=pd.read_csv("/content/Cars.csv")
```

```python
q20.head()
```

```python
# Calculating prob of MPG greater than 38 since we want the cal probability of value greater than 38 we will subtract it from 100 which is 1
mpg_grt_38=np.round(1-stats.norm.cdf(38,loc=q20.MPG.mean(),scale=q20.MPG.std()),3)
print("MPG greater than 38:",mpg_grt_38)
```

```python
# Porbability of MPG less than 40, since it is less than we will not subtract from 1
mpg_les_40=np.round(stats.norm.cdf(40,loc=q20.MPG.mean(),scale=q20.MPG.std()),3)
print("MPG Less than 40:",mpg_les_40)
```

c. P (20<MPG<50)
**This statement demands finding a prob of MPG greater than 20 and less than 50, so will calculate the probability of greater than 20 by 1-stats.norm.cdf and less than 50 by stats.norm.cf and after calculate the difference by subtracting 50-20.**

```python
## similarly value greater than 20 we will subtract it from 100 which is 1 Probability of MPG greater than 20
mpg_grt_20=np.round(1-stats.norm.cdf(20,loc=q20.MPG.mean(),scale=q20.MPG.std()),3)
print("MPG greater than 20:",mpg_grt_20)
```

```python
## Now Pobablitity of MPDG less than 50
mpg_les_50=np.round(stats.norm.cdf(50,loc=q20.MPG.mean(),scale=q20.MPG.std()),3)
print("MPG Greater than 50:",mpg_les_50)
```

```python
## now calculating the difference between greater than 20 and less than 50

mpg_grt_20_les_thn_50=(mpg_les_50)-(mpg_grt_20)
print("probability of MPG greater than 20 and less 50:",mpg_grt_20_les_thn_50)
```

**Q 22- *Calculate* the Z scores of 90% confidence interval,94% confidence interval, 60% confidence interval**

**Answer**:To calculate the Z score we use ppf(percent point function) we are rounding the value till 4 digit



**Explanation on 0.95 derived from,  - Since we want a 90% CI, the remaining 10% (100% - 90%) is divided equally between the two tails of the distribution (5% each)**
## so 90+05 = 95 , which is 0.95**

```python
print("Z scores of 90% confidence  interval is :",np.round(stats.norm.ppf(0.95),4))
```

```python
print("Z scores of94% confidence interval is:",np.round(stats.norm.ppf(0.97),4)) # similary we have derived 0.97 by dividing remaining %
```

```python
print("Z scores of60% confidence  interval is:",np.round(stats.norm.ppf(0.80),4))  # similary we have derived 0.80 by dividing remaining %
```

**Q23 Calculate the t scores of 95% confidence interval, 96% confidence interval, 99% confidence interval for sample size of 25**

Answer- 0.975 derived from, - Since we want a 95% CI, the remaining 05% (100% - 05%) is divided equally between the two tails of the distribution (2.5% each), which is 97.5 =, 0.975, df is Degree of freedom (n-1).

```python
print("T score of 95% CI:",np.round(stats.t.ppf(0.975,df=24),3))
```

```python
print("T score of 96% CT:",np.round(stats.t.ppf(0.98,df=24),3))
```

```python
print("T score of 99% CI:",np.round(stats.t.ppf(0.995,df=24),3))
```
