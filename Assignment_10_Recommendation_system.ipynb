# **Recommendation System**

```python
# Importing libraries
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
# Importing dataset

books=pd.read_csv("/content/books.csv",encoding="latin-1")
books.head()
```

**EDA**

```python
# Droping unnamed column

books.drop(["Unnamed: 0"],axis=1,inplace=True)
```

```python
# Renaming the clomn

books.rename({"User.ID":"ID","Book.Title":"Title","Book.Rating":"Rating"},axis=1,inplace=True)
books.head()
```

```python
# Checking the unique value in the title column

len(books["Title"].unique())
```

```python
# Cheking the unique user ID
len(books["ID"].unique())
```

```python
# Cheking the count of ratings from 1-10

books["Rating"].value_counts()
```

```python
# Creating a pivot table to change the orentation of unique title from row to coloumn

books1=books.pivot_table(index="ID",columns="Title",values="Rating")
books1
```

```python
## Imputing NA to 0
books1.fillna(0,inplace=True)
books1.head()
```

```python
#Calculating Cosine Similarity between Users

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
# User similarity - pairwise - taking 2 rows at a time it will compute distance
# That distance is scaled between 0 to 1

sim=1-pairwise_distances(books1.values,metric="cosine")
sim
```

```python
## Above we can see the diagonal more similarity because because it is comparing 1=1, 2=2
#Imputing digonal values to 0

np.fill_diagonal(sim,0)
sim
```

```python
sim1=pd.DataFrame(sim)
sim1
```

```python
# computing the index and column names to user ids

sim1.index=books.ID.unique()
sim1.columns=books.ID.unique()
sim1
```

```python
# idxmax gives the most Similar Users
sim1.idxmax(axis=1)
```

```python
# Now that it shows that 276736  and   276726 are most similar values we can check

books[(books["ID"]==276736)|(books["ID"]==276726)]
```

```python
#Considering the most similar customer

def recommend(x):
  simid=sim1.loc[x].idxmax()
  simID=books[books["ID"]==simid].Title
  xid=books[books["ID"]==x].Title
  return set(simID)-set(xid)
```

```python
recommend(8)
```
